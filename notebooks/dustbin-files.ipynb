dustbin files are not labeled, but in the krakenbraken file there are assignments of labels.

```python
import pathlib
import pandas as pd
```

```python
path="/data/bacteria/File1_full_krakenbracken.txt"
krakenbraken = pd.read_csv(path,sep="\t")
```

```python
dustbin_files = pathlib.Path("/data/bacteria/experiments/autoencoders/6mer/fcgr/").rglob("dustbin*/*npy")
```

```python
id_samples = [p.stem for p in dustbin_files]
```

```python
krakenbraken.query(f"sample_id in {id_samples}").groupby("V2").size().sort_values()[-30:].plot(kind="barh", figsize=(10,8))
```

```python
krakenbraken.query(f"sample_id in {id_samples}").groupby("V2").size().sort_values()[:30].plot(kind="barh", figsize=(10,8))
```

```python
counts_by_species = krakenbraken.query(f"sample_id in {id_samples}").groupby("V2").size().to_dict()
```

```python
one_representative = [sp for sp,count in counts_by_species.items() if count <= 1]
one_representative, len(one_representative)
```
