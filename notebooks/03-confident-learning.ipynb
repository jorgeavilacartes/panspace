```python
import json
import numpy as np 
import pandas as pd 

from cleanlab.filter import find_label_issues

from sklearn.linear_model import LogisticRegression
```

```python
# paths
path_embeddings_train = lambda KFOLD: f"/home/avila/github/panspace-paper/output/7mer/test-autoencoder/cross-validation/binary_crossentropy-relu-sigmoid-{KFOLD}-fold/faiss-embeddings/embeddings.npy"
path_labels_train = lambda KFOLD: f"/home/avila/github/panspace-paper/output/7mer/test-autoencoder/cross-validation/binary_crossentropy-relu-sigmoid-{KFOLD}-fold/faiss-embeddings/labels.json"

path_embeddings_test = lambda KFOLD: f"/home/avila/github/panspace-paper/output/7mer/test-autoencoder/cross-validation/binary_crossentropy-relu-sigmoid-{KFOLD}-fold/test/embeddings.npy"
path_labels_test = lambda KFOLD: f"/home/avila/github/panspace-paper/output/7mer/test-autoencoder/cross-validation/binary_crossentropy-relu-sigmoid-{KFOLD}-fold/faiss-embeddings/labels.json"
```

```python
def run(KFOLD):
    
    X_train = np.load(path_embeddings_train(KFOLD)) 
    with open(path_labels_train(KFOLD)) as fp:
        y_train = json.load(fp)
        y_train = pd.Series(y_train, name="label") 

    unique_labels = y_train.unique()
    unique_labels.sort()
    dict_labels = {label: idx for idx, label in enumerate(unique_labels)}

    X_test = np.load(path_embeddings_test(KFOLD)) 
    with open(path_labels_train(KFOLD)) as fp:
        y_test = json.load(fp)
        y_test = pd.Series(y_test, name="label")

    labels_test = [dict_labels[label] for label in y_test]

    clf = LogisticRegression(random_state=42).fit(X_train, y_train)

    pred_probs = clf.predict_proba(X_test)

    return pred_probs, labels_test
```

```python
pred_probs, labels = run(4)
```
